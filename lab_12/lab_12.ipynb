### Task 2

To perform the transformation, multiply the matrix being transformed by the _transform matrix_. Examples of transformation matrices for performing transformations are listed below:

Rotate by angle $\theta$:
$$
A_\theta = \begin{pmatrix}
\text{cos}\theta & -\text{sin}\theta\\
\text{sin}\theta & \text{cos}\theta
\end{pmatrix}
$$

Stretch along the $0x$ axis:
$$
A_k = \begin{pmatrix}
k & 0\\
0 & 1
\end{pmatrix}
$$

Stretch along the $0y$ axis:
$$
A_k = \begin{pmatrix}
1 & 0\\
0 & k
\end{pmatrix}
$$

Stretching in the proportion $k_x$ along the $0x$ axis and in the proportion $k_y$ along the $0y$ axis:
$$
A_k = \begin{pmatrix}
k_x & 0\\
0 & k_y
\end{pmatrix}
$$

Tilt relative to the $0y$ axis:
$$
A_k = \begin{pmatrix}
1 & k\\
0 & 1
\end{pmatrix}
$$

Tilt relative to the $0x$ axis:
$$
A_k = \begin{pmatrix}
1 & 0\\
k & 1
\end{pmatrix}
$$

### Task 3 (affine transformations)

Affine transformations allow you to perform a _translation_ operation and combine it with other transformation operations.
To perform an affine transformation, change the representation of the point $(x, y) \in R^2$ to $(x, y, 1) \in R^3$. In this way, the transformation matrix reaches 3x3 dimensions. The matrix enabling the translation operation (shift by the $[k_x, k_y]$ vector) then looks as follows:
$$
A_k = \begin{pmatrix}
1 & 0 & k_x \\
0 & 1 & k_y \\
0 & 0 & 1
\end{pmatrix}
$$

The remaining operations are the same as before, e.g. rotation operation:
$$
A_\theta = \begin{pmatrix}
\text{cos}\theta & -\text{sin}\theta & 0\\
\text{sin}\theta & \text{cos}\theta & 0 \\
0 & 0 & 1
\end{pmatrix}
$$
