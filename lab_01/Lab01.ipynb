### Zad 1:

Zaimplementuj funkcję:
$f(x) = \frac{1}{8}x^{4} + 2(x + 0.7)^{3} - \frac{1}{2}x^{2} - 38 x + 1$

```python
from typing import List

def fun1(element_list: List) -> float:
    result = []
    for x in element_list:
        result.append(0.125 * x ** 4 + 2 * (x + 0.7) ** 3 - 0.5 * x ** 2 - 38 * x + 1)
    return result
```

### Zad 2:

Oblicz wartosc funkcji $f(x)$ w przedziale [-20, 20] z krokiem 1 ($\Delta$).



```python
x = [x for x in range(-20, 21, 1)]
y = fun1(x)
```

### Zad 3:

Wyświetl wykres funkcji $f(x)$.

```python
from matplotlib.pyplot import plot, show

plot(x, y)
show()
```

### Zad 4:

Znajdź taką wartość $x$ (w przedziale _[-20, 20]_), dla której funkcja $f(x)$ przyjmuje najmniejszą wartość.

```python
x_index = y.index(min(y))
print(f'x = {x[x_index]}, y = {min(y)}')
```

### Zad 5:

Wyznacz inną liczbę przedziałów (inny krok ($\Delta$)) z zakresu _[-20, 20]_ i sprawdź czy odnajdziesz mniejszą wartość minimalną funkcji $f(x)$ niż w poprzednim zadaniu? Jeżeli tak to podaj wartość $x$ dla której znalazłeś najmniejszą wartość funkcji $f(x)$.

```python
x_float = [x / 10 for x in range(-200, 201, 1)]
y_float = fun1(x_float)

x_float_index = y_float.index(min(y_float))
x_float_min = x_float[x_float_index]
y_float_min = min(y_float)

print(f'x = {x_float_min}, y = {y_float_min}')
```

### Zad 6:

Wyświetl wykres (parametry wykresu dobierz samodzielnie) prezentujący minimum funkcji $f(x)$


```python
plot(x_float, y_float)
plot(x_float_min, y_float_min, 'ro')
show()
```

### Zad 7:

Znajdź taką wartość $x$ (w przedziale _[-20, 20])_, dla której funkcja $f(x)$ przyjmuje najmniejszą wartość.
W tym zadaniu wygeneruj $k$ losowych wartości $x$ (metoda Monte Carlo).
Wyświetl wykres funkcji.

```python
from random import randrange

k = 100
x_random = [randrange(-20, 20) for _ in range(k)]
x_random = sorted(x_random)
y_random = fun1(x_random)

x_random_index = y_random.index(min(y_random))
x_random_min = x_random[x_random_index]
y_random_min = min(y_random)

print(f'x = {x_random_min}, y = {y_random_min}')

plot(x_random, y_random)
plot(x_random_min, y_random_min, 'ro')
show()
```
