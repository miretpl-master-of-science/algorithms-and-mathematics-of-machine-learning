### Task 1:

Implement function
$f(x) = \frac{1}{8}x^{4} + 2(x + 0.7)^{3} - \frac{1}{2}x^{2} - 38 x + 1$

```python
def func(element_list: list) -> list:
    return [0.125 * v ** 4 + 2 * (v + 0.7) ** 3 - 0.5 * v ** 2 - 38 * v + 1 for v in element_list]
```

### Task 2:

Calculate the value of the function $f(x)$ in the interval _[-20, 20]_ with step 1 ($\Delta$).

```python
x = [x for x in range(-20, 21, 1)]
y = func(x)
```

### Task 3:

Display the graph of $f(x)$.

```python
from matplotlib.pyplot import plot, show

plot(x, y)
show()
```

### Task 4:

Find the value of $x$ (in the interval _[-20, 20]_) for which the function $f(x)$ takes the smallest value.

```python
x_idx = y.index(min(y))
print(f'x = {x[x_idx]}, y = {min(y)}')
```

### Task 5:

Determine a different number of intervals (different step ($\Delta$)) from the range _[-20, 20]_ and check if you find a smaller minimum value of the $f(x)$ function than in the previous task? If so, enter the value of $x$ for which you found the smallest value of the function $f(x)$.

```python
x_float = [x / 10 for x in range(-200, 201, 1)]
y_float = func(x_float)

x_float_idx = y_float.index(min(y_float))
x_float_min = x_float[x_float_idx]
y_float_min = min(y_float)

print(f'x = {x_float_min}, y = {y_float_min}')
```

### Task 6:

Display a graph (select the graph parameters yourself) showing the minimum of the $f(x)$ function.

```python
plot(x_float, y_float)
plot(x_float_min, y_float_min, 'ro')
show()
```

### Task 7:

Find the value of $x$ (in the interval _[-20, 20])_ for which the function $f(x)$ takes the smallest value.
In this problem, generate $k$ random values of $x$ (Monte Carlo method).
Display a graph of the function.

```python
from random import randrange

k = 100
x_rand = sorted([randrange(-20, 20) for _ in range(k)])
y_rand = func(x_rand)

x_rand_min = x_rand[y_rand.index(min(y_rand))]
y_rand_min = min(y_rand)

print(f'x = {x_rand_min}, y = {y_rand_min}')

plot(x_rand, y_rand)
plot(x_rand_min, y_rand_min, 'ro')
show()
```
